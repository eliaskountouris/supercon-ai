```python
#H3O5Li4

def toList(string):
  rtn = []
  i=0
  a=0
  for n in range(len(string)):

    if (i==0):
      if (string[a:n].isalpha()):
        if (string[n:n+1].isalpha() == False):
          rtn.append(elements[string[a:n]])
          i=1
          a=n

    if (i==1):
      if (string[a:n].
          isalpha() == False):
        if (string[n:n+1].isalpha()):
          rtn.append(float(string[a:n]))
          i=0
          a=n
        if (n == len(string)-1):
          rtn.append(float(string[a:n+1]))
          i=0
          a=n

  X = np.reshape(rtn, (int(len(rtn) / 2), 2))   
  return X
```
